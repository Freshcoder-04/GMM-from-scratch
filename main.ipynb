```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sys
from scipy.stats import multivariate_normal
from gmm import GMM
```

```python
data = pd.read_csv('data.csv')
```

```python
gmm = GMM(3)
gmm.fit(data.iloc[:,:2].values)
```

```python

# Get the parameters of the model
params = gmm.getParams()
print("Means:", params["means"])
print("Covariances:", params["covariances"])
print("Weights:", params["weights"])

# Get the membership values (responsibilities)
membership = gmm.getMembership(data.iloc[:,:2])
print("Membership values:", membership)

# Get the likelihood of the data under the model
likelihood = gmm.getLikelihood(data.iloc[:,:2])
print("Log-Likelihood:", likelihood)
```

```python
plt.figure()
centers = gmm.means
plt.scatter(data.iloc[:,0],data.iloc[:,1],marker='o',s=10,c='black')
plt.scatter(centers[:,0],centers[:,1],c='red',s=60)
plt.show()
```
